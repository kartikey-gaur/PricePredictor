```python
import numpy as np
import pandas as pd
import math as math

def grad(x, y):
    mc=0
    bc=0
    n= len(x);
    iterate =1000
    learning_rate= .001
    
    for i in range(iterate):
        yp= mc*x+ bc
        md= -(2/n)*sum(x*(y-yp))
        bd= -(2/n)*sum(y-yp)
        mc= mc- learning_rate*md
        bc= bc - learning_rate*bd
        print("m:{}, b:{}, iterate: {}".format(mc, bc, i))

x= np.array([1,2,3,4,5])
y= np.array([5,7,9,11,13])

grad(x,y)
```

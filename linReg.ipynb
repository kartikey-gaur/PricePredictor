```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("price.csv") ;
df
```

```python

%matplotlib inline
plt.xlabel("area");
plt.ylabel("price") ;
plt.scatter(df["area"], df["price"], color="blue", marker="*")
```

```python
#creates an object of linear regression model.
lreg = linear_model.LinearRegression() ;
lreg.fit(df[["area"]], df.price) 

#lreg.predict([[3300]])
#df1 = pd.read_csv("area.csv");
#df1
```

```python
lreg.predict([[3300]])
```

```python
df1 = pd.read_csv("area.csv");
df1
```

```python
p = lreg.predict(df1)
p
```

```python
df1["price"]= p ;
df1
```

```python
df1.to_csv("predict.csv");
```

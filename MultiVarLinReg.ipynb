```python
#import all the needed modules
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import math as math
```

```python
#import the data set in the data frame
#df is data frame which is basically a trwo-dimensional array

df = pd.read_csv("multi.csv")
df
```

```python
#before doing machine learning or training your regression model, make sure that
#our data is correct at all points

#to remove empty cells we can remove the row, or fill that cell with a measure of central tendency.

#we can use median or mean or mode

#here we do median

p= math.floor(df.bed.median())
df.bed = df.bed.fillna(p)
df
```

```python
preg = linear_model.LinearRegression() ;
preg.fit(df[["area", "bed", "old"]], df.price);
preg.coef_
```

```python
preg.intercept_
```

```python
preg.predict([[2989, 5, 3]])
```
